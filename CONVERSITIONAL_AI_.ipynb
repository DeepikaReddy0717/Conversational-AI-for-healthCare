```python
import pandas as pd
from google.colab import files

uploaded = files.upload()
filename = next(iter(uploaded))
df = pd.read_csv(filename, encoding='utf-8')
print(df.columns.tolist())
df.head()
```

```python
df = df.dropna(subset=['Patient','Doctor'])
df = df[['Patient','Doctor']]

def clean_text(text):
    import re, string
    text = str(text).lower()
    text = re.sub(r'[^a-z\s]', '', text)
    return text.strip()

df['Patient_clean'] = df['Patient'].apply(clean_text)
df['Doctor_clean'] = df['Doctor'].apply(clean_text)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(df['Patient_clean'])
```

```python
def get_response(user_input):
    user_input_clean = clean_text(user_input)
    user_vec = tfidf.transform([user_input_clean])
    sim_scores = cosine_similarity(user_vec, tfidf_matrix)
    idx = sim_scores.argmax()
    response = df.iloc[idx]['Doctor']
    return response
```

```python
print("Healthcare Conversational AI â€” type 'exit' to stop")

while True:
    user_input = input("You: ")
    if user_input.lower() in ['exit','quit','stop']:
        print("AI: Take care! ðŸ‘‹")
        break
    response = get_response(user_input)
    print(f"AI: {response}")
```
